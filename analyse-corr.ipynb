```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxon_df = pd.read_csv("/kaggle/input/eau-dataset/données/taxon.csv")
taxon_denombre_df = pd.read_csv("/kaggle/input/eau-dataset/données/taxon_denombre.csv")
liste_df = pd.read_csv("/kaggle/input/eau-dataset/données/liste_faunistique_floristique.csv")
stations_df = pd.read_csv("/kaggle/input/eau-dataset/données/station.csv")
physicochimie_df = pd.read_csv("/kaggle/input/eau-dataset/données/physicochimie.csv")
alteration_df = pd.read_csv("/kaggle/input/eau-dataset/données/classe_alteration_seq_eau.csv")
```

On s'intéresse à l'évolution du nombre d'individus, on isole ainsi les lignes s'intéressant à cette propriété

```python
critere_nombre_individus = taxon_denombre_df[taxon_denombre_df["critere_taxon"] == "Nombre exact d'individus"]\
    ["critere_taxon_id"]\
    .unique()\
    .tolist()
critere_nombre_individus
```

On vérifie la cohérence entre l'id du critère et sa description (l'autre sens a déjà été vérifié par la selection précédente)

```python
df = taxon_denombre_df[taxon_denombre_df["critere_taxon_id"].isin(critere_nombre_individus)]
df["critere_taxon"].unique()
```

On ne s'intéresse qu'à un unique critère. On a vérifié la cohérence de notre séléction : les colonnes propres aux critères sont inutiles

Ensuite, on remarque qu'une seule mesure est bien effectuée par taxon par liste ff

```python
individus_taxons = df.drop(columns=["critere_taxon", "critere_taxon_id"])

individus_taxons\
    .groupby(["taxon_id", 'liste_ff_id'])\
    .count()\
    ["valeur"]\
    .unique()
```

```python
mesure_par_taxon = individus_taxons\
    .drop(columns=["valeur"])\
    .groupby(["taxon_id"])\
    .count()\
    ["liste_ff_id"]

taxons_to_keep = mesure_par_taxon[mesure_par_taxon > 21000]\
    .index\
    .tolist()
taxons_to_keep
```

```python
test_indiv = individus_taxons[individus_taxons["taxon_id"].isin(taxons_to_keep)]
test_indiv
```

```python
liste_df
```

```python
liste_df_to_merge = liste_df.drop(columns=["date_prel", "facies_vitesse", "facies_profondeur", "facies_substrat"])
stn_lf = stations_df\
    .drop(columns=["nom_masse_eau", "st_y", "st_x", "date_year"])\
    .merge(liste_df_to_merge, left_on='id', right_on='id')
stn_taxon = stn_lf.merge(test_indiv, left_on='liste_faunistique_floristique', right_on='liste_ff_id').drop(columns=["liste_faunistique_floristique"])
stn_taxon.drop_duplicates(inplace=True)
stn_taxon.sort_values(["taxon_id", "liste_ff_id"])
```

```python
pivot_params = physicochimie_df.pivot(index=None, columns="parameter", values="valeur")
params_year_stn = pd\
    .concat([physicochimie_df[["stn_id", "date_year"]], pivot_params], axis=1)\
    .groupby(["stn_id", "date_year"])\
    .first()\
    .reset_index()
params_year_stn
```

```python
stn_taxon_pc = stn_taxon.merge(params_year_stn, left_on=["id", "date_year"], right_on=["stn_id","date_year"])
result = stn_taxon_pc.drop(columns=["id"])
corr_df = result[(result["her"] == 22) & (result["taxon_id"] == 807)]
```

```python
station_count = result[result["taxon_id"] == 807]["stn_id"].value_counts()
station = station_count.index[0]
station_num = station_count.iloc[0]


corr_df = result[(result["stn_id"] == station) & (result["taxon_id"] == 807)]
corr = corr_df.drop(columns=["her", "date_year", "liste_ff_id", "taxon_id", "stn_id"]).corr()
x = corr[["valeur"]]
x
```

```python
x_tot = {}
for taxon in taxons_to_keep:
    x_tot[taxon] = None
    for k in range(10):
        station_count = result[result["taxon_id"] == taxon]["stn_id"].value_counts()
        station = station_count.index[k]
        station_num = station_count.iloc[k]


        corr_df = result[(result["stn_id"] == station) & (result["taxon_id"] == taxon)]
        corr = corr_df.drop(columns=["her", "date_year", "liste_ff_id", "taxon_id", "stn_id"]).corr()
        x = corr[["valeur"]]
        x_tot[taxon] = x if x_tot[taxon] is None else x_tot[taxon] + x
        plt.figure(figsize=(10,5))
        hm = sns.heatmap(x, annot=True)
        hm.set(xlabel='Correlation value', ylabel='Parameters', title = f"Correlation taxon {taxon}, station : {station}, relevés : {station_num}")
    plt.show()

    her_count = result[result["taxon_id"] == taxon]["her"].value_counts()
    her = her_count.index[0]
    her_num = her_count.iloc[0]
    corr_df = result[(result["her"] == her) & (result["taxon_id"] == taxon)]
    corr = corr_df.drop(columns=["her", "date_year", "liste_ff_id", "taxon_id", "stn_id"]).corr()
    x = corr[["valeur"]]
    plt.figure(figsize=(10,5))
    hm = sns.heatmap(x, annot=True)
    hm.set(xlabel='Correlation value', ylabel='Parameters', title = f"Correlation taxon {taxon}, her : {her}, relevés : {her_num}")

    plt.show()
```

On observe que de nombreux taxons sont corrélés avec le paramètre 1348 sur certaines stations, on peut poser l'hypothèse que l'évolution de ce paramètre impacte ces taxons. De plus on observe également le taxon 933 est corrélé avec plusieurs paramètres sur certaines stations. Il pourrait être plus sensible que les autres taxons à des paramètres physico-chimique

```python
for taxon in taxons_to_keep:
    plt.figure()
    hm = sns.heatmap(x_tot[taxon].divide(10), annot=True)
    hm.set(xlabel='Correlation value', ylabel='Parameters', title = f"Correlation taxon {taxon}, x_tot")
plt.show()
```

```python
temp1 = stations_df[(stations_df['date_year'] == 2013)]
temp2 = temp1.merge(liste_df, left_on='id', right_on='id')
temp3 = temp2.merge(taxon_denombre_df, left_on='liste_faunistique_floristique', right_on='liste_ff_id')
taxon_faune_nbr = temp3[temp3['critere_taxon_id'] == 1]
```

```python
ph_test = physicochimie_df[physicochimie_df['stn_id'].isin(taxon_faune_nbr['id'])]
```

```python
def plot_graph(taxon_faune_nbr, ph_test, station_test, taxon_test, p):
    
    df_station_test = taxon_faune_nbr[taxon_faune_nbr['id'] == station_test]
    
    fig, ax = plt.subplots(figsize = (12,6))
    sns.lineplot(data=df_station_test[(df_station_test['taxon_id'] == taxon_test)], x="date_year_y", y="valeur",
                 hue="taxon_id", ci=None, ax=ax, palette="tab10")
    fig.autofmt_xdate()
    plt.title('{} evolution on station {}'.format(taxon_test, station_test))
    plt.show()
    ph_test_stn = ph_test[ph_test['stn_id'] == station_test]
    fig, ax = plt.subplots(figsize = (12,6))
    sns.lineplot(data=ph_test_stn[(ph_test_stn['parameter'] == p)], x="date_year", y="valeur", hue='parameter_name', ci=None, ax=ax, palette="tab10")
    fig.autofmt_xdate()
    plt.title('{} evolution on station {}'.format(p, station_test))
    plt.show()
```

```python
station_test = 405076
taxon_test = 212
p = 1348


plot_graph(taxon_faune_nbr, ph_test, station_test, taxon_test, p)
```

```python
station_test = 403697
taxon_test = 212
p = 1348


plot_graph(taxon_faune_nbr, ph_test, station_test, taxon_test, p)
```

La corrélation trouvé sur la silice sur les nombreux taxons est en fait du au manque de mesure et n'est donc pas utilisabel

```python
station_test = 404759
taxon_test = 212
p = 1350


plot_graph(taxon_faune_nbr, ph_test, station_test, taxon_test, p)
```
